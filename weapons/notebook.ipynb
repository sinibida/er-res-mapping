```python
l10n_english_path = './l10n_english.txt'

weapon_names = []

pattern = re.compile(r"Item/Name/(\d+)┃(?:<color=\w+>)?([^<]+)(?:</color>)?")
with open(l10n_english_path) as f:
  for line in f:
    line = line.strip()
    if line.startswith("Item/Name/"):
      match = pattern.match(line)

      if match is None:
        continue

      weapon_names.append({
          "code": int(match.group(1)),
          "engName": match.group(2),
      })

df_weapon_names = pd.DataFrame(weapon_names);

df_weapon_names
```

```python
df_weapon_list = df_weapon_list.merge(df_weapon_names, left_on='code', right_on='code')
df_weapon_list
```

```python
# 예외 engName -> 실제 파일 이름
# LATER: not ok 전부 해결
eng_name_replace = {
    "Sharpened Spear": "Shapened Spear",
    "Long Sword": "Long_Sword",
    "Twin Blades": "Twin Swords",
    "Black Butterfly": "Black Butterfly",
    "Lloigor & Zahr": "Lloigor & Zahr",
    "Eighteen Foot Spear": "Eighteen Foor Spear",
    "Cathode Lash": "Cathode Lash",
    "Imperial Silk Gloves": "Imperial Silk Gloves",
    "Ryukyu Tonfa": "Ryukyu Tonfa",
    "Stampede": "Stampede",
    "Fencing Rapier": "Fencing Rapier",
    "Teen Spirit": "Teen Spirit",
    "Instant Camera": "Instant Camera",
    "Call of Cadmus Lv2": "Call of Cadmus Lv2",
    "Call of Cadmus Lv3": "Call of Cadmus Lv3",
    "Alpha Sidewinder ML": "Alpha Sidewinder ML",
    "Alpha Sidewinder ML": "Alpha Sidewinder ML",
}


def is_match(engName: str, path: pathlib.Path):
  # # 디버깅용
  # if engName == "Sharpened Spear":
  #   print(engName, path.stem)

  if engName in eng_name_replace:
    engName = eng_name_replace[engName]

  # not ok: Scarlet Dagger - Crimson
  # not ok: Scarlet Dagger - Dawn
  if engName.endswith(" - Crimson"):
    engName = engName.replace(" - Crimson", "")
  if engName.endswith(" - Dawn"):
    engName = engName.replace(" - Dawn", "")

  engName = engName.replace("'", "_")

  return engName in path.stem


def create_pair(code, path):
  return {
      'code': code,
      'path': path,
  }


root = "01. Weapons"
ok_cnt = 0
not_ok_cnt = 0
weapon_map = []
for (index, row_weapon) in df_weapon_list.iterrows():
  ok = False
  weapon_type_dir = weapon_type_to_directory[row_weapon['weaponType']]
  weapon_type_path = pathlib.Path(root, weapon_type_dir)

  for file in weapon_type_path.iterdir():
    if is_match(row_weapon['engName'], file):
      path = str(file)
      weapon_map.append(create_pair(row_weapon['code'], path))
      ok = True
      break

  if ok:
    ok_cnt += 1
  else:
    print("not ok:", row_weapon['engName'])
    not_ok_cnt += 1

df_weapon_map = pd.DataFrame(weapon_map)

print("OK:", ok_cnt)
print("NO:", not_ok_cnt)
df_weapon_map
```

```python
output = {}
code_prefix = "weapon/"
path_prefix = "/Eternal Return Fankit/Item/"
for i, row in df_weapon_map.iterrows():
  output[code_prefix + str(row['code'])] = path_prefix + row['path']

with open("output.json", "w") as f:
  json.dump(output, f)
```
