```python
import json
import pandas as pd
import re
import pathlib
```

```python
character_skin_json_path = "./CharacterSkin.json"

with open(character_skin_json_path) as f:
  obj = json.load(f)

obj
```

```python
character_skin = obj['data']

df_character_skin = pd.DataFrame(character_skin)
df_character_skin.sort_values("code", ignore_index=True, inplace=True)
df_character_skin
```

```python
df_character_skin[df_character_skin['index'] == 0]
```

```python
root = "Character"

df_character_list = df_character_skin[df_character_skin['index'] == 0]

root_path = pathlib.Path(root)

character_path_map = {}

for dir in root_path.iterdir():
  folder = dir.name
  for i, row in df_character_list.iterrows():
    if row['name'] in folder:
      character_path_map[row['characterCode']] = folder

character_path_map
```

```python

for (index, row_skin) in df_character_skin.iterrows():
  print(index, row_skin)
  if (index > 10):
    break;
```

```python
# 예외 engName -> 실제 파일 이름
eng_name_replace = {
}

name_pattern = re.compile(r"\d+\.(.+)")
def is_match(name: str, path: pathlib.Path, index: int):
  # # 디버깅용
  # if engName == "Sharpened Spear":
  #   print(engName, path.stem)

  name_pattern_match = name_pattern.match(path.name)
  if name_pattern_match is None:
    print("OH SHIT:", path.name)
    return False
  
  path_name: str = name_pattern_match.group(1)

  if index == 0:
    return path_name.strip().lower() == "default"

  return path_name.strip().lower().startswith(name.lower())


def create_pair(code, path):
  return {
      'code': code,
      'path': path,
  }


ok_cnt = 0
not_ok_cnt = 0
skin_map = []

for (index, row_skin) in df_character_skin.iterrows():
  # For Debugging
  if row_skin['characterCode'] not in [1, 2, 7]:
    continue

  ok = False
  character_dir = character_path_map[row_skin['characterCode']]
  character_path = pathlib.Path(root, character_dir)

  for file in character_path.iterdir():
    if is_match(row_skin['name'], file, row_skin['index']):
      path = str(file)
      skin_map.append(create_pair(row_skin['code'], path))
      ok = True
      break

  if ok:
    ok_cnt += 1
  else:
    print("not ok:", row_skin['name'])
    not_ok_cnt += 1

df_skin_map = pd.DataFrame(skin_map)

print("OK:", ok_cnt)
print("NO:", not_ok_cnt)
df_skin_map
```

```python
output = {}
code_prefix = "skin/"
path_prefix = "/Eternal Return Fankit/"
for i, row in df_skin_map.iterrows():
  output[code_prefix + str(row['code'])] = path_prefix + row['path']

with open("output.json", "w") as f:
  json.dump(output, f)
```
